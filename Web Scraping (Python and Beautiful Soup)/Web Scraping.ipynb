### More CODEWITHHARRY

```python
import requests
from bs4 import BeautifulSoup

url='https://www.codewithharry.com'

#get the html
r=requests.get(url)
htmlContent=r.content
# print(htmlContent)

# PARSE the HTML  (creating SOUP)
soup=BeautifulSoup(htmlContent,'html.parser')

#HTML tree traversal
title=soup.title  # webpage ko tile dinxa

print(type(title))   # (1) tag
print(type(title.String))   # (2)  NavigableString
print(type(soup))    # (3) Beautifulsoup
```

```python
# GET FIRST PARAGRAPH 

first_para=soup.find('p')

print(first_para)


## GET ALL THE PARAGRAPH FROM THE PAGE

all_paras=soup.find_all('p')
print(all_paras)
```

```python
# GET Classes of any element in the HTML page

all_classes=soup.find(('p'),['class'])
print(all_classes)


# find all elements with particular class ( all elements with class name 'lead')
lead_class=soup.find_all('p',class_="lead")
print(lead_class)
```

```python
## GET the TExT from  the TAGS/ SOUP

#get text from soup
get_all_text_from_soup=soup.get_text()
print(get_all_text_from_soup)

#get text from tags

get_text_from_par_tag=soup.find("p").get_text()
print(get_text_from_par_tag)
```

```python
# GET ALL ANCHOR TAGS

anchors=soup.find_all('a')
print(anchors)
```

```python
## GET ALL LINKS FROM ANCHOR TAG.

anchors=soup.find_all('a')
all_links=set()
for link in anchors:
    if(link.get('href') !='#'):
        linkText="http://codewithharry.com"+ link.get('href')
        all_links.add(link)
        print(linkText)
```

```python

## COMMENT OBJECT 

markup=" <p> <!-- this is comment --> </p>"
soup2=BeautifulSoup(markup)
print(soup2.p.string)
print(type(soup2.p.string))
```

```python
## Give element with particular ID
navbarSupport=soup.find(id='navbarSupportedContent')
print(navbarSupport.children)  #gives all the children of that elements
print(navbarSupport.contents)  # gives all contents of that elements in the form of tlist (use loopto iterate over list)
```

```python
# Gives all the contents inside that element like( Div vitra from hunxa paragaraph hunxa etc)

for elem in navbarSupport.children:
    print(elem)
    
    
for elem in navbarSupport.contents:
    print(elem)
```

```python
# gives all the string inside particular elements

for item in navbarSupport.strings:
    print(item)
    

# scattered string in one place 

for item in navbarSupport.stripped_strings:
    print(item)
```

```python
## Get  PARENTS 

print(navbarSupport.parent) # gives the parents of the particular tag


# parents - Gives all the parents above the children
# parent - Gives only one parents name 

for item in navbarSupport.children:
    print(item.name)
```

```python
## Getting  SIBLINGS 

print(navbarSupport.next_sibling) #gives next siblings

print(navbarSupport.next_sibling.next_sibling) # next ko pani next dinxa

print(navbarSupport.previous_sibling) # gives previous siblings 

print(navbarSupport.previous_sibling.previous_sibling)

```
